## Detrending a signal

```python
import numpy as np
from scipy import signal
from matplotlib import pyplot as plt
```

Generate a random signal with a trend

```python
t = np.linspace(0, 5, 100)
data = t + np.random.normal(size=100)
```

Detrend

```python
from scipy import signal
detrended_data = signal.detrend(data, type='linear')
```

Plot

```python
plt.figure(figsize=(10, 8))
plt.plot(t, data, label="data")
plt.plot(t, detrended_data, label="detrended data")
plt.xlabel('x')
plt.ylabel('y')
plt.title('Example of Detrending a Dataset')
plt.savefig('Example of Detrending a Dataset.jpg')
plt.show()
```
