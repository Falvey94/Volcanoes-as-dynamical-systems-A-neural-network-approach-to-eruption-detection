```python
White_Island_data = pd.read_csv('WhiteIsland_2019_data.csv')

#White_Island_data[(White_Island_data['Data'] < 0.0002) & (White_Island_data['Data'] > -0.0002)].dropna()

White_Island_data.index = White_Island_data['Times'] #make Time the index row of the dataframe
White_Island_data.pop('Times') #drop the original times column
White_Island_data=White_Island_data[100:50300:]

#White_Island_data_adj = White_Island_data[(White_Island_data[43280:43350:]<0.0002) & (White_Island_data[43280:43350:]>-0.0002)].dropna()
#White_Island_data_adj = White_Island_data[43280:43350:]
White_Island_data_adj = White_Island_data[43310:43350:]

Velocity = White_Island_data_adj['Data']
Velocity.plot(ylabel = 'Velocity (m/s)')

x2t_2delay = np.array(White_Island_data_adj['Data'])

#White_Island_data_adj2 = White_Island_data[43285:43355]
White_Island_data_adj2 = White_Island_data[43315:43355]
x2t_1delay = np.array(White_Island_data_adj2['Data'])

#White_Island_data_adj3 = White_Island_data[43290:43360:]
White_Island_data_adj3 = White_Island_data[43320:43360:]
x2t = np.array(White_Island_data_adj3['Data'])
```

```python
#3d plot

plt.figure(figsize=(10, 8), dpi=80)

ax = plt.axes(projection='3d')

ax.plot3D(x2t, x2t_1delay, x2t_2delay, 'blue',label= 'Active State')
ax.plot3D(xt, xt_1delay, xt_2delay, 'red', label= 'Rest State')

ax.set_xlabel('x(t)', fontsize =14)
ax.set_ylabel('x(t-)',fontsize =14)
ax.set_zlabel('y(t-2)',fontsize =14)
ax.set_title('3D view of attractors for Processed Data',fontsize =18)

ax.legend()

ax.figure.savefig('Processed_Data_Atractor_Visual.png')
```
