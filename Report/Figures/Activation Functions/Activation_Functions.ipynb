#  Activation Functions

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#BinaryStep

def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)

#tanh

def tanh(x):
    ''' It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between -1 to 1.'''
    return np.tanh(x)

#Sigmoid

def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''
    return 1/(1+np.exp(-x))

#ReLU

def RELU(x):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1

#SeLU

def SELU(x, lambdaa = 1.0507, alpha = 1.6732):
    ''' It returns zero if the input is less than zero otherwise it returns the given input. '''
    x1=[]
    for i in x:
        if i<0:            
            x1.append(lambdaa * alpha * (np.exp(i) - 1))
        else:            
            x1.append(lambdaa * i)
            
    return x1
```

```python
x = np.linspace(-5, 5)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function: Binary Step')
plt.grid()
plt.show()
```

```python
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function: Sigmoid')
plt.grid()
plt.show()
```

```python
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function: tanh')
plt.grid()
plt.xlim(-5,5)
plt.show()
```

```python
plt.plot(x, RELU(x))
plt.axis('tight')
plt.title('Activation Function: ReLU')
plt.grid()
plt.xlim(-5,5)
#plt.ylim(-1,1)
plt.show()
```

```python
x = np.linspace(-5, 5)

plt.plot(x, SELU(x))
plt.axis('tight')
plt.title('Activation Function: Selu')
plt.grid()
plt.xlim(-5,5)
plt.ylim(-2,5)
plt.show()
```

```python
plt.figure(figsize=(16,8))
plt.suptitle('Activation Functions', fontsize=18)

x = np.linspace(-6, 6)

plt.subplot(2,2,1), plt.plot(x, RELU(x),  color = 'blue')
plt.title('ReLU'), plt.grid(), plt.xlim(-5,5)

plt.subplot(2,2,2), plt.plot(x, SELU(x), color = 'green')
plt.title('SeLU'), plt.grid(), plt.xlim(-5,5)

plt.subplot(2,2,3), plt.plot(x, sigmoid(x), color = 'red')
plt.title('Sigmoid'), plt.grid(), plt.xlim(-5,5)

plt.subplot(2,2,4), plt.plot(x, tanh(x), color = 'black')
plt.title('Tanh'), plt.grid(), plt.xlim(-5,5)

plt.show()
plt.savefig('Activation_Functions.png')
```

```python
fig = plt.figure(figsize=(16,8))
fig.suptitle('Activation Functions', fontsize=18)

x = np.linspace(-6, 6)

ax1 = fig.add_subplot(2, 2, 1)
ax1.plot(x, RELU(x),  color = 'blue')
ax1.set_title('ReLU') , ax1.grid(), ax1.set_xlim(-5,5)

ax2 = fig.add_subplot(2, 2, 2)
ax2.plot(x, SELU(x), color = 'green')
ax2.set_title('SeLU'), ax2.grid(), ax2.set_xlim(-5,5)

ax3 = fig.add_subplot(2, 2, 3)
ax3.plot(x, sigmoid(x), color = 'red')
ax3.set_title('Sigmoid'), ax3.grid(),ax3.set_xlim(-5,5)

ax4 = fig.add_subplot(2, 2, 4)
ax4.plot(x, tanh(x), color = 'black')
ax4.set_title('Tanh'), ax4.grid(), ax4.set_xlim(-5,5)

fig.show()
fig.savefig('Activation_Functions.png')
```
